```python
import pandas as pd
import numpy as np

with open('input.txt', 'r') as archivo:
    filas = []
    for linea in archivo:
        numeros = [float(numero) for numero in linea.strip().split('  ')]
        numeros1 = numeros
        len_numeros =len(numeros)
        len_numeros1 = len_numeros
        M = []
        for i in numeros:
           for j in numeros1:
              M.append(np.absolute(i-j))
        Mn = np.reshape(M, (len_numeros, len_numeros1))
        Mn_mean = np.mean(Mn)
        Mn_max= np.max(Mn)
        numeros.append(Mn_mean)
        numeros.append(Mn_max)
        filas.append(numeros)

        result = np.where(Mn == np.amax(Mn))
        listOfCordinates = list(zip(result[0], result[1]))
        for cord in listOfCordinates:
            print(cord)
        numeros.append(cord)
        print(Mn)
        print(numeros)
with open('output.txt', 'w') as archivo_nuevo:
    for fila in filas:
        linea = ' '.join(str(numero) for numero in fila)
        linea += f' {" "}\n'
        archivo_nuevo.write(linea)

Python Version: 3.10.12
Pandas Version: 2.0.3
Numpy Version: 1.25.2

Appendix C

Python Script for Variational AutoEncoder Neural Network

#Libraries

import keras
from keras.layers import Conv2D, Conv2DTranspose
from keras.layers import Dense, Input, Flatten, Lambda, Reshape
from keras.models import Model
from keras import backend as K
import numpy as np
import matplotlib.pyplot as plt
import csv
import sys


np.set_printoptions(threshold=sys.maxsize)


# %%

#Load Images
with open('/home/ev3th/Tesis/Imagenes/Image1.csv', 'r') as f:
    reader = csv.reader(f)
    Image1 = list(reader)

Image1_array = np.array(Image1, dtype=float)
Image1_array1 = Image1_array.reshape((-1, 1))
Image1_array1 = np.reshape(Image1_array1, (50, 50, 1))
Image1_array1 = Image1_array1.astype('uint8')




with open('/home/ev3th/Tesis/Imagenes/Image2.csv', 'r') as f:
    reader = csv.reader(f)
    Image2 = list(reader)

Image2_array = np.array(Image2, dtype=float)
Image2_array2 = Image2_array.reshape((-1, 1))
Image2_array2 = np.reshape(Image2_array2, (50, 50, 1))
Image2_array2 = Image2_array2.astype('uint8')




with open('/home/ev3th/Tesis/Imagenes/Image3.csv', 'r') as f:
    reader = csv.reader(f)
    Image3 = list(reader)

Image3_array = np.array(Image3, dtype=float)
Image3_array3 = Image3_array.reshape((-1, 1))
Image3_array3 = np.reshape(Image3_array3, (50, 50, 1))
Image3_array3 = Image3_array3.astype('uint8')




with open('/home/ev3th/Tesis/Imagenes/Image4.csv', 'r') as f:
    reader = csv.reader(f)
    Image4 = list(reader)

Image4_array = np.array(Image4, dtype=float)
Image4_array4 = Image4_array.reshape((-1, 1))
Image4_array4 = np.reshape(Image4_array4, (50, 50, 1))
Image4_array4 = Image4_array4.astype('uint8')



with open('/home/ev3th/Tesis/Imagenes/Image5.csv', 'r') as f:
    reader = csv.reader(f)
    Image5 = list(reader)

Image5_array = np.array(Image5, dtype=float)
Image5_array5 = Image5_array.reshape((-1, 1))
Image5_array5 = np.reshape(Image5_array5, (50, 50, 1))
Image5_array5 = Image5_array5.astype('uint8')



with open('/home/ev3th/Tesis/Imagenes/Image6.csv', 'r') as f:
    reader = csv.reader(f)
    Image6 = list(reader)

Image6_array = np.array(Image6, dtype=float)
Image6_array6 = Image6_array.reshape((-1, 1))
Image6_array6 = np.reshape(Image6_array6, (50, 50, 1))
Image6_array6 = Image6_array6.astype('uint8')




with open('/home/ev3th/Tesis/Imagenes/Image7.csv', 'r') as f:
    reader = csv.reader(f)
    Image7 = list(reader)

Image7_array = np.array(Image7, dtype=float)
Image7_array7 = Image7_array.reshape((-1, 1))
Image7_array7 = np.reshape(Image7_array7, (50, 50, 1))
Image7_array7 = Image7_array7.astype('uint8')



with open('/home/ev3th/Tesis/Imagenes/Image8.csv', 'r') as f:
    reader = csv.reader(f)
    Image8 = list(reader)

Image8_array = np.array(Image8, dtype=float)
Image8_array8 = Image8_array.reshape((-1, 1))
Image8_array8 = np.reshape(Image8_array8, (50, 50, 1))
Image8_array8 = Image8_array8.astype('uint8')



with open('/home/ev3th/Tesis/Imagenes/Image9.csv', 'r') as f:
    reader = csv.reader(f)
    Image9 = list(reader)

Image9_array = np.array(Image9, dtype=float)
Image9_array9 = Image9_array.reshape((-1, 1))
Image9_array9 = np.reshape(Image9_array9, (50, 50, 1))
Image9_array9 = Image9_array9.astype('uint8')



with open('/home/ev3th/Tesis/Imagenes/Image10.csv', 'r') as f:
    reader = csv.reader(f)
    Image10 = list(reader)

Image10_array = np.array(Image10, dtype=float)
Image10_array10 = Image10_array.reshape((-1, 1))
Image10_array10 = np.reshape(Image10_array10, (50, 50, 1))
Image10_array10 = Image10_array10.astype('uint8')



with open('/home/ev3th/Tesis/Imagenes/Image11.csv', 'r') as f:
    reader = csv.reader(f)
    Image11 = list(reader)

Image11_array = np.array(Image11, dtype=float)
Image11_array11 = Image11_array.reshape((-1, 1))
Image11_array11 = np.reshape(Image11_array11, (50, 50, 1))
Image11_array11 = Image11_array11.astype('uint8')



with open('/home/ev3th/Tesis/Imagenes/Image12.csv', 'r') as f:
    reader = csv.reader(f)
    Image12 = list(reader)

Image12_array = np.array(Image12, dtype=float)
Image12_array12 = Image12_array.reshape((-1, 1))
Image12_array12 = np.reshape(Image12_array12, (50, 50, 1))
Image12_array12 = Image12_array12.astype('uint8')


with open('/home/ev3th/Tesis/Imagenes/Image13.csv', 'r') as f:
    reader = csv.reader(f)
    Image13 = list(reader)

Image13_array = np.array(Image13, dtype=float)
Image13_array13 = Image13_array.reshape((-1, 1))
Image13_array13 = np.reshape(Image13_array13, (50, 50, 1))
Image13_array13 = Image13_array13.astype('uint8')


with open('/home/ev3th/Tesis/Imagenes/Image14.csv', 'r') as f:
    reader = csv.reader(f)
    Image14 = list(reader)

Image14_array = np.array(Image14, dtype=float)
Image14_array14 = Image14_array.reshape((-1, 1))
Image14_array14 = np.reshape(Image14_array14, (50, 50, 1))
Image14_array14 = Image14_array14.astype('uint8')



with open('/home/ev3th/Tesis/Imagenes/Image15.csv', 'r') as f:
    reader = csv.reader(f)
    Image15 = list(reader)

Image15_array = np.array(Image15, dtype=float)
Image15_array15 = Image15_array.reshape((-1, 1))
Image15_array15 = np.reshape(Image15_array15, (50, 50, 1))
Image15_array15 = Image15_array15.astype('uint8')



with open('/home/ev3th/Tesis/Imagenes/Image16.csv', 'r') as f:
    reader = csv.reader(f)
    Image16 = list(reader)

Image16_array = np.array(Image16, dtype=float)
Image16_array16 = Image16_array.reshape((-1, 1))
Image16_array16 = np.reshape(Image16_array16, (50, 50, 1))
Image15_array16 = Image16_array16.astype('uint8')



with open('/home/ev3th/Tesis/Imagenes/Image17.csv', 'r') as f:
    reader = csv.reader(f)
    Image17 = list(reader)

Image17_array = np.array(Image17, dtype=float)
Image17_array17 = Image17_array.reshape((-1, 1))
Image17_array17 = np.reshape(Image17_array17, (50, 50, 1))
Image17_array17 = Image17_array17.astype('uint8')



with open('/home/ev3th/Tesis/Imagenes/Image18.csv', 'r') as f:
    reader = csv.reader(f)
    Image18 = list(reader)

Image18_array = np.array(Image18, dtype=float)
Image18_array18 = Image18_array.reshape((-1, 1))
Image18_array18 = np.reshape(Image18_array18, (50, 50, 1))
Image18_array18 = Image18_array18.astype('uint8')



with open('/home/ev3th/Tesis/Imagenes/Image19.csv', 'r') as f:
    reader = csv.reader(f)
    Image19 = list(reader)

Image19_array = np.array(Image19, dtype=float)
Image19_array19 = Image19_array.reshape((-1, 1))
Image19_array19 = np.reshape(Image19_array19, (50, 50, 1))
Image19_array19 = Image19_array19.astype('uint8')



with open('/home/ev3th/Tesis/Imagenes/Image20.csv', 'r') as f:
    reader = csv.reader(f)
    Image20 = list(reader)

Image20_array = np.array(Image20, dtype=float)
Image20_array20 = Image20_array.reshape((-1, 1))
Image20_array20 = np.reshape(Image20_array20, (50, 50, 1))
Image20_array20 = Image20_array20.astype('uint8')



with open('/home/ev3th/Tesis/Imagenes/Image21.csv', 'r') as f:
    reader = csv.reader(f)
    Image21 = list(reader)

Image21_array = np.array(Image21, dtype=float)
Image21_array21 = Image21_array.reshape((-1, 1))
Image21_array21 = np.reshape(Image21_array21, (50, 50, 1))
Image21_array21 = Image21_array21.astype('uint8')



with open('/home/ev3th/Tesis/Imagenes/Image22.csv', 'r') as f:
    reader = csv.reader(f)
    Image22 = list(reader)

Image22_array = np.array(Image22, dtype=float)
Image22_array22 = Image22_array.reshape((-1, 1))
Image22_array22 = np.reshape(Image22_array22, (50, 50, 1))
Image22_array22 = Image22_array22.astype('uint8')



with open('/home/ev3th/Tesis/Imagenes/Image23.csv', 'r') as f:
    reader = csv.reader(f)
    Image23 = list(reader)

Image23_array = np.array(Image23, dtype=float)
Image23_array23 = Image23_array.reshape((-1, 1))
Image23_array23 = np.reshape(Image23_array23, (50, 50, 1))
Image23_array23 = Image23_array23.astype('uint8')



with open('/home/ev3th/Tesis/Imagenes/Image24.csv', 'r') as f:
    reader = csv.reader(f)
    Image24 = list(reader)

Image24_array = np.array(Image24, dtype=float)
Image24_array24 = Image24_array.reshape((-1, 1))
Image24_array24 = np.reshape(Image24_array24, (50, 50, 1))
Image21_array24 = Image24_array24.astype('uint8')



peces = np.stack((Image1_array1, Image2_array2, Image3_array3, Image4_array4, Image5_array5, Image6_array6, Image7_array7, Image8_array8, Image9_array9, Image10_array10, Image11_array11, Image12_array12, Image13_array13, Image14_array14, Image15_array15, Image16_array16, Image17_array17, Image18_array18, Image19_array19, Image20_array20, Image21_array21, Image22_array22, Image23_array23, Image24_array24))

x_train = peces.astype('float32')

x_test = peces.astype('float32')


# %%


with open('/home/ev3th/Tesis/y_test.csv', 'r') as f:
    reader = csv.reader(f)
    y_test = list(reader)

y_test_array = np.array(y_test)
y_test_array = np.reshape(y_test_array, (24))

y_test = y_test_array.astype('uint8')
y_train = y_test

# %%

# Reshape
img_width = x_train.shape[1]
img_height = x_train.shape[2]
num_channels = 1


x_train = x_train.reshape(x_train.shape[0], img_height, img_width, num_channels)
x_test = x_test.reshape(x_test.shape[0], img_height, img_width, num_channels)
input_shape = (img_height, img_width, num_channels)



# %%

# Image#1 (Sample of 4 images before entering the CVAE)
fig, axs = plt.subplots(2, 2, figsize=(15, 15))

# First subplot
axs[0, 0].imshow(Image1_array, cmap='copper', interpolation='nearest')
axs[0, 0].set_title('Image 1')
fig.colorbar(axs[0, 0].images[0], ax=axs[0, 0])

# Second subplot
axs[1, 0].imshow(Image7_array, cmap='copper', interpolation='nearest')
axs[1, 0].set_title('Image 7')
fig.colorbar(axs[1, 0].images[0], ax=axs[1, 0])

# Third subplot
axs[0, 1].imshow(Image15_array, cmap='copper', interpolation='nearest')
axs[0, 1].set_title('Image 15')
fig.colorbar(axs[0, 1].images[0], ax=axs[0, 1])

# Fourth subplot
axs[1, 1].imshow(Image22_array, cmap='copper', interpolation='nearest')
axs[1, 1].set_title('Image 22')
fig.colorbar(axs[1, 1].images[0], ax=axs[1, 1])


plt.tight_layout()
plt.show()



# %%

# Dimensionality Reduction
latent_dim = 2

input_img = Input(shape=input_shape, name= 'encoder_input')
x = Conv2D(32, 3, padding='same', activation='relu')(input_img)
x = Conv2D(64, 3, padding='same', activation='relu', strides=(2,2))(x)
x = Conv2D(64, 3, padding='same', activation='relu')(x)
x = Conv2D(64, 3, padding='same', activation='relu')(x)

conv_shape = K.int_shape(x)

x = Flatten()(x)
x = Dense(32, activation='relu')(x)

# %%

z_mu = Dense(latent_dim, name = 'latent_mu')(x)
z_sigma = Dense(latent_dim, name='latent_sigma')(x)

def sample_z(args):
    z_mu, z_sigma = args
    eps = K.random_normal(shape=(K.shape(z_mu)[0], K.int_shape(z_mu)[1]))
    return z_mu + K.exp(z_sigma / 2) *eps


z = Lambda(sample_z, output_shape=(latent_dim, ), name='z')([z_mu, z_sigma])


#%%

encoder = Model(input_img, [z_mu, z_sigma, z], name ='encoder')
print(encoder.summary())

#%%

decoder_input = Input(shape=(latent_dim, ), name = 'decoder_input')

x = Dense(conv_shape[1]*conv_shape[2]*conv_shape[3], activation='relu')(decoder_input)
x = Reshape((conv_shape[1], conv_shape[2], conv_shape[3]))(x)

x = Conv2DTranspose(32, 3, padding='same', activation='relu', strides=(2,2))(x)

x = Conv2DTranspose(num_channels, 3, padding='same', activation='sigmoid', name='decoder_output')(x)


#%%

decoder = Model(decoder_input, x, name='decoder')
decoder.summary()

#%%

z_decoded = decoder(z)

#%%

class CustomLayer(keras.layers.Layer):
    def vae_loss(self, x, z_decoded):
        x = K.flatten(x)
        z_decoded = K.flatten(z_decoded)

        recon_loss = keras.metrics.binary_crossentropy(x, z_decoded)

        kl_loss = -5e-4 * K.mean(1 + z_sigma - K.square(z_mu) - K.exp(z_sigma), axis = -1)
        return K.mean(recon_loss + kl_loss)
    def call(self, inputs):
        x = inputs[0]
        z_decoded = inputs[1]
        loss = self.vae_loss(x, z_decoded)
        self.add_loss(loss, inputs = inputs)
        return x


y = CustomLayer()([input_img, z_decoded])


vae = Model(input_img, y, name='vae')

vae.compile(optimizer = 'adam', loss = None)
vae.summary()

vae.fit(x_train, None, epochs = 100, batch_size = 6, validation_split = 0.2)


#%%

# Image#2 (Dimensionality reduction)
mu, _, _ = encoder.predict(x_test)
# Representacion de la distribuxion mu como

x = mu[:, 0]
y = mu[:, 1]



z_muscle = []
for i in range (1,13):
    z_muscle.append("#" + str(i) + "m")

z_stomach = []
for j in range (13,25):
    z_stomach.append("#" + str(j) + "s")

z_ms = (z_muscle + z_stomach)

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot()

plt.scatter(x, y, c=y_test, cmap='winter', s=50)
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.colorbar()

for j, txt in enumerate(z_ms):
    ax.annotate(txt, (x[j], y[j]))

plt.show()



#%%

# Imagen#3 (New synthetic Image creation)
sample_vector = np.array([[1, 0]])
decoded_example = decoder.predict(sample_vector)
decoded_example_reshaped = decoded_example.reshape(img_width, img_height)
plt.figure(figsize = (15, 15))
plt.imshow(decoded_example_reshaped, cmap= 'copper', interpolation = 'nearest')
plt.colorbar()
plt.show()



#%%

# Image#4 (New synthetic creation of a 5x5 Images Matrix)

n = 5 #generate 15x15 digits
figure = np.zeros((img_width * n, img_height * n, num_channels))

grid_x = np.linspace(-5, 5, n)
grid_y = np.linspace(-5, 5, n)[::1]

#decoder for each square in the grid
for i, yi in enumerate(grid_y):
    for j, xi in enumerate(grid_x):
        z_sample = np.array([[xi, yi]])
        x_decoded = decoder.predict(z_sample)
        digit = x_decoded[0].reshape(img_width, img_height, num_channels)
        figure[ i * img_width: (i + 1) * img_width, j * img_height: (j + 1) * img_height] = digit


plt.figure(figsize = (10, 10))
#Reshape for visualization
fig_shape = np.shape(figure)
figure = figure.reshape((fig_shape[0], fig_shape[1]))

#plt.imshow(figure, cmap='gnuplot2')
plt.imshow(figure, cmap= 'copper', interpolation = 'nearest')
plt.show()
```
